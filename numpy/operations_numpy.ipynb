```python
import numpy as np

```

```python
#operations on arrays

arr = np.array([[1,2,3],[4,5,6]])
print(arr, arr.shape)

reshaped = arr.reshape(3,2)
print(reshaped, reshaped.shape)

flattened = arr.flatten() #2D>1D
print(flattened, flattened.shape)
```

```python
#indexing 
arr = np.array([1,2,3,4,5])
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr[0])
print(arr[3])
print(arr2[0][2])


```

```python
#fancy indexing

arr = np.array([1,2,3,4,5])

idx = [0,1,3]
print(arr[idx])
```

```python
#boolean indexing

arr = np.array([1,2,3,4,5,])

print(arr[arr > 3])
print(arr[arr%2==0])
print(arr[arr%2!=0])
```

```python
#slicing

arr= np.array([1,2,3,4,5])

print(arr[1:4])
print(arr[1:])
print(arr[:4])
print(arr[::2])
```

```python
 # python lib vs numpy
# lib makes copy but numpy makes view(original)

nums = [1,2,3,4,5] 
sub_list = nums[1:3]
print(sub_list)

sub_list[0] = 200
print(sub_list)
print(nums)


arr = np.array([1,2,3,4,5])
sub_arr = arr[1:3]
print(sub_arr)

sub_arr[0] = 200
print(sub_arr)
print(arr)
```

```python
# to make copy from the array using numpy

arr = np.array([1,2,3,4,5])
sub_arr = arr[1:3].copy() #no change will reflect in array
print(sub_arr)

sub_arr[0] = 200
print(sub_arr)
print(arr)
```
